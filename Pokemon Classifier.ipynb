## Data Processing

```python
import os
from pathlib import Path
from keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
p = Path("./dataset/")
dirs = p.glob('*')

image_data = []
labels = []

image_paths = []

label_dict = {'Pikachu':0, 'Bulbasaur':1, 'Meowth':2}
pokemon_label = {0:'Pikachu', 1:'Bulbasaur', 2:'Meowth'}

for folder_dir in dirs:
    label = str(folder_dir).split('\\')[-1]
    
    count=0
    print(folder_dir)
    
    for img_path in folder_dir.glob('*.jpg'):
        img = image.load_img(img_path, target_size=(100,100))
        img_array = image.img_to_array(img)
        image_data.append(img_array)
        
        labels.append(label_dict[label])
        
        count += 1
    
    print(count)
```

## Visualization

```python
import numpy as np
```

```python
X = np.array(image_data)
Y = np.array(labels)

print(X.shape, Y.shape)
```

```python
def drawImage(img, label):
    plt.title(pokemon_label[label])
    plt.axis('off')
    plt.imshow(img)
    plt.show()
```

```python
for i in range(10):
    rand_index = np.random.randint(400)
    drawImage(X[rand_index]/255.0, Y[rand_index])
```
